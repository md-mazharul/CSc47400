```python
# ---------- Problem 2 ----------
import numpy as np
import matplotlib.pyplot as plt

start_period = "1994-1998"
end_period   = "2007-2008"
```

```python
men_fruit_pct = ((df_men_fruit.loc[end_period] - df_men_fruit.loc[start_period])
                 / df_men_fruit.loc[start_period].replace(0, np.nan)) * 100
men_fruit_pct = men_fruit_pct.dropna().sort_values()

print("Men – Fruit % change")
print(men_fruit_pct.round(1))
```

```python
men_dairy_pct = ((df_men_dairy.loc[end_period] - df_men_dairy.loc[start_period])
                 / df_men_dairy.loc[start_period].replace(0, np.nan)) * 100
men_dairy_pct = men_dairy_pct.dropna().sort_values()

print("Men – Dairy % change")
print(men_dairy_pct.round(1))
```

```python
gap = 1
x_fruit = np.arange(len(men_fruit_pct))
x_dairy = np.arange(len(men_dairy_pct)) + len(men_fruit_pct) + gap

fig, ax = plt.subplots(figsize=(12,5))
ax.bar(x_fruit, men_fruit_pct.values)
ax.bar(x_dairy, men_dairy_pct.values)

xticks = list(men_fruit_pct.index) + list(men_dairy_pct.index)
ax.set_xticks(np.concatenate([x_fruit, x_dairy]))
ax.set_xticklabels(xticks, rotation=45, ha="right")

ax.set_title("Percent Change by Product (Men): 1994–1998 → 2007–2008")
ax.set_ylabel("Percent change (%)")
ax.axhline(0, color="black", lw=1)

ax.text(x_fruit.mean(), ax.get_ylim()[1]*0.95, "Fruit", ha="center", va="top")
ax.text(x_dairy.mean(), ax.get_ylim()[1]*0.95, "Dairy", ha="center", va="top")

plt.tight_layout()
plt.show()
```
